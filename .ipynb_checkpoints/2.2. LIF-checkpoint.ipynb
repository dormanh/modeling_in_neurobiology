## Exercise 1.: minimal current

In the absence of an input current, a LIF neuron has a constant membrane
voltage `V_REST`. If an input current drives `vm` above the firing
threshold, a spike is generated. Then, `vm` is reset to `V_RESET` and
the neuron ignores any input during the refractroy period.

### Question 1.1: minimal current (calculation)

For the default neuron parameters (see above), compute the minimal
amplitude $I_{min}$ of a step current to elicitate a spike. You can
access the default values in your code and do the calculation with
correct units:

### Answer 1.1

**Based on equation [(1.7)](https://neuronaldynamics.epfl.ch/online/Ch1.S3.html#Ch1.E7), the asymptotic value of the membrane potential (the value it approaches in case of an indefinitely long input) is $u(\infty) = u_{rest} + RI_0$. Thus, the minimal input current required to elicit a spike is such that: $I_{min} = min(\{I_0: u_{rest} + RI_0 \geq \vartheta\})$ (where $\vartheta$ is the firing threshold).**

```python
I_min = (LIF.FIRING_THRESHOLD - LIF.V_REST) / LIF.MEMBRANE_RESISTANCE
```

```python
print(f"The minimal required input current is: {round(float(I_min) * 10 ** 9)} nA")
```

### Question 1.2: minimal current (simulation)

Use the value $I_{min}$ you've computed and verify your result: inject a
step current of amplitude $I_{min}$ for 100ms into the LIF neuron and
plot the membrane voltage. `vm` should approach the firing threshold but
**not** fire. We have implemented a couple of helper functions to solve
this task. Use this code block, but make sure you understand it and
you've read the docs of the functions `.LIF.simulate_LIF_neuron`,
`.input_factory.get_step_current` and
`.plot_tools.plot_voltage_and_current_traces`.

```python
step_current = input_factory.get_step_current(
    t_start=5, t_end=105, unit_time=b2.ms, amplitude=I_min
)
```

```python
state_monitor, spike_monitor = LIF.simulate_LIF_neuron(
    input_current=step_current, simulation_time=120 * b2.ms
)
```

```python
plot_tools.plot_voltage_and_current_traces(
    state_monitor,
    step_current,
    title="min input",
    firing_threshold=LIF.FIRING_THRESHOLD,
)
print("nr of spikes: {}".format(spike_monitor.count[0]))
```

## Exercise 2: f-I Curve

For a constant input current $I$, a LIF neuron fires regularly with
firing frequency $f$. If the current is to small ($I < I_{min}$) $f$ is
0Hz; for larger $I$ the rate increases. A neuron's firing-rate versus
input-amplitude relationship is visualized in an "f-I curve".

### Question 2.1: f-I Curve and refractoryness

We now study the f-I curve for a neuron with a refractory period of 3 ms
(see `.LIF.simulate_LIF_neuron` to learn how to set a refractory
period).

1.  Sketch the f-I curve you expect to see.
2.  What is the maximum rate at which this neuron can fire?
3.  Inject currents of different amplitudes (from 0nA to 100nA) into a
    LIF neuron. For each current, run the simulation for 500ms and
    determine the firing frequency in Hz. Then plot the f-I curve. Pay
    attention to the low input current.

### Answer 2.1
- **The resulting firing frequency is supposed to be an increasing, but concave function of the input current amplitude (since it has an upper bound).**
- **The maximum rate at which the neuron can fire is determined by the refractory period: $\frac{1}{3} ms \sim 333.33 Hz$.**
- **See below the simulation for 11 different input current amplitudes from 1 to 100 nA.**

```python
input_amplitudes = np.concatenate([np.arange(10), np.linspace(10, 100, 10)])[1:]
```

```python
step_currents = [
    input_factory.get_step_current(
        t_start=5, t_end=105, unit_time=b2.ms, amplitude=I * b2.nA
    )
    for I in input_amplitudes
]
```

```python
monitors = np.array(
    [
        LIF.simulate_LIF_neuron(
            input_current=step_current,
            simulation_time=500 * b2.ms,
            abs_refractory_period=3.0 * b2.ms,
        )
        for step_current in tqdm(step_currents)
    ]
)
```

**Here you can interactively modify the input current amplitude and see the resulting firing pattern.**

```python
@widgets.interact
def plot_spikes(i=(0, input_amplitudes.shape[0] - 1, 1)):

    plot_tools.plot_voltage_and_current_traces(
        monitors[i, 0],
        step_currents[i],
        title=f"input: {input_amplitudes[i]} nA",
        firing_threshold=LIF.FIRING_THRESHOLD,
    )
    print("nr of spikes: {}".format(monitors[i, 1].count[0]))
```

**The firing frequency is calculated as the reciprocal of the interspike interval.**

```python
spike_times = [
    np.array(monitors[:, 1][t].values(var="t")[0])
    for t in range(input_amplitudes.shape[0])
]
```

```python
frequencies = [0, 0] + [
    1 / (spike_time[1] - spike_time[0]) for spike_time in spike_times[2:]
]
```

```python
plt.plot(input_amplitudes, frequencies, c="orange", lw=3)
plt.xlabel("Input current amplitude (nA)", fontsize="large")
plt.ylabel("Firing frequency (Hz)", fontsize="large")
plt.xticks(input_amplitudes[9:])
plt.ylim(0, 350)
plt.title("f-I curve", fontsize="x-large")
```

**Here is a close up to lower amplitudes.**

```python
plt.plot(input_amplitudes[:10], frequencies[:10], c="orange", lw=3)
plt.xlabel("Input current amplitude (nA)", fontsize="large")
plt.ylabel("Firing frequency (Hz)", fontsize="large")
plt.xticks(input_amplitudes[:10])
plt.ylim(0, 350)
plt.title("f-I curve", fontsize="x-large")
```

### Answer 3.

- **resting potential: $\sim -63 mV$ (there's a constant line at this value before the input onset)**
- **reset voltage: $\sim -60 mV$ (this is the value the voltage returns to after the spike)**
- **firing threshold: $\sim -38 mV$ (gradually modifying the input current amplitude reveals that in case at least one spike is elicited, the peak of the membrane voltage curve is around this value)**
- **absolute refractory period: $\sim 4.5 ms$ (testing large input current amplitudes reveals that it is not possible to elicit repetitive firing with a larger frequency)**

**Based on the figure, the minimal input current amplitude required to elicit a spike - given that the input duration is maximal - is $\sim 4.6 mV$ (however, I can't be sure, as the simulation only runs only for 50 ms). The membrane resistance can be estimated based on the asymptotic solution of equation (1.7) and the estimated firing threshold, resting potential and minimal required input current: $\frac{-38 mV + 63 mV}{4.6 nA} = \sim 5.4 M ohm$.**

```python
R = (-38 * b2.mV + 63 * b2.mV) / (4.6 * b2.nA)
```

```python
R
```

**For the membrane time scale, I use equation (1.7) again, according to which the membrane potential evolves as $u(t) = u_{rest} + RI_0 (1 - e^{-\frac{t}{\tau_m}})$. The only unknown variable here is $\tau_m$, which I can infere from the observation. Solving for $\tau_m$, we get: $\tau_m = -\frac{t}{ln(1 - \frac{u(t) - u_{rest}}{RI_0})}$. Below, I inject a $10 nA$ input current for 10 milliseconds to the neuron and observe the evolution of the membrane potential in the relevant time period. Based on this and the known parameters, I can estimate $\tau_m$.**

```python
u_rest = -63 * b2.mV
I_0 = 10 * b2.namp
```

```python
test_current = input_factory.get_step_current(
    t_start=5, t_end=15, unit_time=b2.ms, amplitude=I_0,
)

test_state_monitor, test_spike_monitor = LIF.simulate_random_neuron(
    test_current, random_parameters,
)
```

```python
relevant_u_ts = np.array(test_state_monitor.v)[0, 51:148] * 1000 * b2.mV
```

```python
ts = np.arange(relevant_u_ts.shape[0]) / 10 * b2.ms
```

```python
tau_m = (-ts / np.log((R * I_0 - relevant_u_ts + u_rest) / (R * I_0))).mean()
```

## Exercise 4: Sinusoidal input current and subthreshold response

In the subthreshold regime (no spike), the LIF neuron is a linear system
and the membrane voltage is a filtered version of the input current. In
this exercise we study the properties of this linear system when it gets
a sinusoidal stimulus.

### Question 4.1

Create a sinusoidal input current (see example below) and inject it into
the LIF neuron. Determine the phase and amplitude of the membrane
voltage.

```python
sinusoidal_current = input_factory.get_sinusoidal_current(
    200,
    1000,
    unit_time=0.1 * b2.ms,
    amplitude=2.5 * b2.namp,
    frequency=250 * b2.Hz,
    direct_current=0.0 * b2.namp,
)
```

```python
sin_state_monitor, sin_spike_monitor = LIF.simulate_LIF_neuron(
    input_current=sinusoidal_current,
    simulation_time=120 * b2.ms,
    firing_threshold=70 * b2.mV,
    v_rest=0 * b2.mV,
    v_reset=5 * b2.mV,
)
```

```python
plot_tools.plot_voltage_and_current_traces(
    sin_state_monitor, sinusoidal_current, title="Sinusoidal input current"
)
print("nr of spikes: {}".format(spike_monitor.count[0]))
```

### Answer 4.1

***Note: I know this task could be solved empirically, but I tried to implement the theoretical model instead.***

**The input here the the following sinewave: $I_0 e^{i \omega t}$ (where $I_0$ is the amplitude and $\omega$ is the frequency of the wave). Based on equation ([1.30](https://neuronaldynamics.epfl.ch/online/Ch1.S3.html#Ch1.E30)), the amplitude of the corresponding output is: $u_0 = |\hat{\kappa}(\omega)| I_0$, where $\kappa(s)$ is the causal linear filter: $\frac{1}{C} e^{-\frac{s}{\tau_m}}$.**

**$|\hat{\kappa}(\omega)|$ - which is the absolute value of the Fourier-transform of $\kappa(s)$ - is calculated as: $|\frac{1}{C} \int_0^{\infty} e^{-t(\frac{1}{\tau_m} + i\omega)} dt| = \frac{1}{C} |\frac{\tau_m}{1 + i \omega \tau_m}|$. Here, $C = \frac{\tau_m}{R}$ and $R$ is the membrane resistance, and $\tau_m$ is the membrane time scale, so the previous term can be rewritten as: $|\frac{R}{1 + i \omega \tau_m}|$.**

```python
def response_amplitude(input_freq, input_amp=2.5):

    return np.abs(
        LIF.MEMBRANE_RESISTANCE
        / (1 + 1j * (input_freq * b2.Hz) * LIF.MEMBRANE_TIME_SCALE)
    ) * (input_amp * b2.namp)
```

```python
response_amplitude(250)
```
