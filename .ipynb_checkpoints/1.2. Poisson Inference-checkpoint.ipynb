# Section 1: The Poisson Distribution

Bayesian analysis operates on probability distributions. Although these can take many forms, the Poisson distribution is a very common choice because of its simplicity: it expresses the probability of a given number of events occurring in a fixed interval of time (or space) if these events occur with a known constant mean rate and independently of the time since the last event.

In this exercise, you will implement a Poisson distribution by filling in the missing portion of `my_poisson` below. Poisson distributions have one parameters: The **mean** $\lambda$, which sets both its average count (mean rate) and its spread around the average: in the case of the Poisson distribution the **variance** equals its mean. On one hand, this gives the Poisson distribution its simplicity: we only need to know the mean rate and we know everything about the process. On the other hand, this is a severe limitation: the variance is tightly coupled to the mean.

The equation for a Poisson distribution is:
$$
\text{Pois}(\lambda) = \frac{\lambda^k\,e^{-\lambda}}{k!}
$$
We perform the calculations in the log domain since computers don't like to deal with very large numbers ($e^k$ and $k!$ can be large when $k\sim20$). At the end we take the exponential of the numbers. The log of the Poisson is:
$$
\ln\big(\text{Pois}(\lambda)\big) = k\,\ln(\lambda) -\lambda - \ln(k!)
$$
Also, don't forget that this is a probability distribution and should therefore sum to one. While this happens "automatically" when summed from $0$ to $\infty$, your version will only be computed over a finite number of points. You therefore need to explicitly normalize it yourself. 

Test out your implementation with a $\lambda = 2$. After you have it working, play with the  parameters to develop an intuition for how changing $\lambda$ alter the shape of the Poisson distribution.

## Exercise 1: Implement a Poisson (2p)

Complete the function defining the Poisson distribution below


```python
def my_poisson(k_spikes, rate):
    """
    Returns normalized Poisson evaluated at points `k_spikes`, with parameter
     mean `rate` 

    Args:
        k_spikes (numpy array of floats): points at which the Poisson is evaluated
        rate (scalar): mean of the Poisson

    Returns:
        (numpy array of floats) : normalized Poisson evaluated at `k_spikes`
    """

    px = np.exp(
        [
            k * np.log(rate) - rate
            if (k == 0)
            else k * (np.log(rate) - np.log(k) + 1) - rate
            for k in k_spikes
        ]
    )

    return px / px.sum()
```

```python
counts = np.arange(0, 20, dtype=float)
px = my_poisson(counts, 4.3)
my_plot_pois(counts, px)
```

## Exercise 2: implement a Poisson likelihood (2p).
Complete the function defining the Poisson *likelihood* below! 

```python
def poisson_likelihood(k_spikes, rates, normalise=False):
    """
    Returns the Poisson likelihood of different rates, given the number of spikes observed

    Args:
        k_spikes (scalar): number of spikes for which the likelihoods are evaluated
        rates (1D numpy array of floats): means of the Poisson for which the likelihood
        is calculated

    Returns:
        (numpy array of floats) : Poisson likelihood of rates, given `k_spikes`
    """

    likelihoods = []

    for rate in rates:

        if rate == 0:
            likelihoods.append(1 if (k_spikes == 0) else 0)

        else:
            likelihoods.append(
                np.exp(
                    k_spikes * np.log(rate) - rate
                    if (k_spikes == 0)
                    else k_spikes * (np.log(rate) - np.log(k_spikes) + 1) - rate
                )
            )

    likelihoods = np.array(likelihoods)

    if normalise:
        return likelihoods / likelihoods.sum()

    return likelihoods
```

```python
rates = np.linspace(0, 20, 201)
likelihood = poisson_likelihood(2, rates, normalise=True)
my_plot_likelihood(rates, likelihood)
```

# Section 2. Bayes' Theorem and the Posterior


Imagine an experiment where the animal (fish) has to decide between two different stimuli based on the spikes emitted by a given neuron - a *kingfisher detector cell*. To decode the object identity (swallow or kingfisher), the participants can use two sources of information: 
  1. new noisy sensory information (the likelihood)
  2. prior expectations of the frequency of the particular objects. 

The prior and the observation are both uncertain, so the subject will need to combine these sources of information to better estimate the identity of the object. 

Bayes' rule tells us how to combine two sources of information: the prior (e.g., a noisy representation of our expectations about where the stimulus might come from) and the likelihood (e.g., a noisy representation of the stimulus position on a given trial), to obtain a posterior distribution taking into account both pieces of information. Bayes' rule states:

\begin{eqnarray}
\text{Posterior} = \frac{ \text{Likelihood} \times \text{Prior}}{ \text{Normalization constant}}
\end{eqnarray}

When the prior is categorical (either kingfisher or swallow is present with corresponding probabilities) and likelihood is Poisson, this translates into the following form:

$$
\begin{array}{rcl}
\text{Likelihood: } P(\text{spikes}|\text{swallow}) &=& \text{Pois}(k; \lambda_\text{swallow}) \\
\text{Prior: }P(\text{swallow}) &=& \alpha \\
\text{Posterior: }P( \text{swallow} | \text{spikes} )&\propto& \alpha\, \text{Pois}(k; \lambda_\text{swallow})\\
P( \text{kingfisher} | \text{spikes} )&\propto& (1-\alpha)\, \text{Pois}(k; \lambda_\text{kingfisher})\\
\end{array}
$$


In Exercise 2, we will use this form of the posterior, where the two distributions are combined via pointwise multiplication.  Although this method requires more computation, it works for any type of probability distribution. 

## Exercise 2: Finding the posterior computationally (3p)


We will use Poisson distributions to represent the likelihood (orange), and a Bernoulli distribution to represent prior expectation (blue). Since there are only two options (swallow or kingfisher) we evaluate the likelihood only at the firing rate values corresponding to these options. Using Bayes rule, you will combine them into a posterior distribution that summarizes the probability that the object is swallow or kingfisher. 

We have provided you with a ready-to-use plotting function, and a code skeleton.

* Use `poisson_likelihood`, the answer to exercise 1, to evaluate the likelihood with parameters $\lambda_{sw}$ = 5 and $\lambda_{kf}$ = 10 and observation $k=9$ spikes.
* Start with an uninformative prior distribution with $P(\text{swallow}) = \alpha = 0.5$ and $P(\text{kingfisher}) = 1-\alpha = 0.5$ 
* Calculate the posterior using pointwise multiplication of the likelihood and prior.
* Plot the likelihood, prior and posterior using the predefined function `posterior_plot`



```python
def compute_posterior_pointwise(prior, likelihood):

    posterior = prior * likelihood

    return posterior / np.sum(posterior)


likelihood_birds = poisson_likelihood(9, np.array([5, 10]))
alpha = 0.5
prior_birds = np.array([alpha, 1 - alpha])
posterior_birds = compute_posterior_pointwise(prior_birds, likelihood_birds)
posterior_plot(prior_birds, likelihood_birds, posterior_birds)
```

Explore the effect of the prior parameter $\alpha$ into the effect of the inference. What is the value $\alpha$ that changes the results?

**Solution 1.: using an approximation**

```python
alphas = np.linspace(0, 1, 100)
swallow_posterior = np.array(
    [
        compute_posterior_pointwise(np.array([alpha, 1 - alpha]), likelihood_birds)[0]
        for alpha in alphas
    ]
)
```

```python
matplotlib.rcParams["figure.figsize"] = (10, 10)
plt.plot(alphas, swallow_posterior, label="swallow posterior")
plt.xlabel("alpha", fontsize="x-large")
plt.ylabel("p(swallow)", fontsize="x-large")
plt.title(
    "Posterior probability of seeing a swallow as a function of alpha",
    fontsize="x-large",
)
threshold = alphas[np.where(swallow_posterior > 0.5)[0][0]]
plt.hlines(0.5, 0, 1, colors="green", label="p(swallow) = 0.5")
plt.vlines(threshold, 0, 1, colors="red", label="threshold alpha")
plt.legend(fontsize="x-large")
```

```python
print(f"The result changes, if alpha is greater than ~{round(threshold, 4)}")
```

**Solution 2.: using the exact formula**

$p(swallow | \alpha, spikes=9) = \frac{\alpha \cdot p(spikes=9 | swallow)}{\alpha \cdot p(spikes=9 | swallow) + (1 - \alpha) \cdot p(spikes=9 | kingfisher)}$

$p(kingfisher | \alpha, spikes=9) = \frac{(1 - \alpha) \cdot p(spikes=9 | kingfisher)}{\alpha \cdot p(spikes=9 | swallow) + (1 - \alpha) \cdot p(spikes=9 | kingfisher)}$

We are looking for the $\alpha$, for which $p(swallow | \alpha, spikes=9) = p(kingfisher | \alpha, spikes=9)$:

$\frac{\alpha \cdot p(spikes=9 | swallow)}{\alpha \cdot p(spikes=9 | swallow) + (1 - \alpha) \cdot p(spikes=9 | kingfisher)} = \frac{(1 - \alpha) \cdot p(spikes=9 | kingfisher)}{\alpha \cdot p(spikes=9 | swallow) + (1 - \alpha) \cdot p(spikes=9 | kingfisher)}$

$\alpha \cdot p(spikes=9 | swallow) = (1 - \alpha) \cdot p(spikes=9 | kingfisher)$

$(p(spikes=9 | swallow) + p(spikes=9 | kingfisher)) \cdot \alpha = p(spikes=9 | kingfisher)$

$\alpha = \frac{p(spikes=9 | kingfisher)}{p(spikes=9 | swallow) + p(spikes=9 | kingfisher)}$

```python
print(
    f"The result changes, if alpha is greater than {round(likelihood_birds[1] / likelihood_birds.sum(), 4)}"
)
```

## Bonus (16p)

This is a standard inference problem but I put it in a different form for the computational neuroscience course:

Two animals appeared in the scene. One elicited a response quite common in the population ('010', 60% of animals elicit this response) while the other triggered a rather unusual answer: '111' - only 1% of the animals trigger this spike combination. A suspected predator, the kingfisher, is tested and found to elicit the usual combination of spikes: '010'. Do these data (type '010' and ‚Äò111‚Äô spike patterns recorded) give evidence in favour of the proposition that kingfisher was one of the two animals present at the scene?

Hint: What is your data? What is the prior information? Formalize the hypothesis and the alternative hypothesis and calculate the probability of the data under the two alternative hypotheses.

**Solution:**

$p('010') = 60 \%$

$p('111') = 1 \%$

$p('010'|kingfisher) = 1$

**The question is, whether the conditional probability given the evidence is greater than the prior probability without the evidence that one of the animals was a kingfisher. Let $p(kingfisher) = \alpha$ denote the relative frequency of kingfishers among all animals.**

**When two birds are observed, the following combinations are possible (the respective probabilites are also indicated):**
- not kingfisher + not kingfisher $\Rightarrow (1 - \alpha)^2$
- **kingfisher + not kingfisher** $\Rightarrow \alpha(1 - \alpha)$
- **not kingfisher + kingfisher** $\Rightarrow (1 - \alpha) \alpha$
- **kingfisher + kingfisher** $\Rightarrow \alpha^2$

$p(\textit{at least one kingfisher}) = 2 \alpha (1 - \alpha) + \alpha^2 = 2 \alpha - \alpha^2$. **So this is our prior probability without further evidence.**

---------------

**When we also measure the elicited responses, the combined possibilities of birds + responses with the respective probabilites are:**
- **not kingfisher + not kingfisher + ['010', '111']** $\Rightarrow (1 - \alpha)^2 \cdot 0.6 \cdot 0.01 = 0.006 (1 - \alpha)^2$
- not kingfisher + not kingfisher + different response $\Rightarrow (1 - \alpha)^2 \cdot (1 - 0.6 \cdot 0.01) = 0.994 (1 - \alpha)^2$
- **kingfisher + not kingfisher + ['010', '111']** $\Rightarrow \alpha(1 - \alpha) \cdot 1 \cdot 0.01 = 0.01 \alpha(1 - \alpha)$
- kingfisher + not kingfisher + different response $\Rightarrow \alpha(1 - \alpha) \cdot 1 \cdot 0.99 = 0.99 \alpha(1 - \alpha)$
- not kingfisher + kingfisher + ['010', '111'] $\Rightarrow (1 - \alpha) \alpha \cdot 0.6 \cdot 0 = 0$
- not kingfisher + kingfisher + different response $\Rightarrow (1 - \alpha) \alpha$
- kingfisher + kingfisher + ['010', '111'] $\Rightarrow \alpha^2 \cdot 1 \cdot 0 = 0$
- kingfisher + kingfisher + different response $\Rightarrow \alpha^2$

**The cases with zero probabilities, as well as cases in which a different response is elicited are excluded, which leaves us with the two cases highlighted with a bold font style. Of these, only one contains a kingfisher.**

$p(\textit{at least one kingfisher} | ['010', '111']) = \frac{0.01 \alpha (1 - \alpha)}{0.01 \alpha (1 - \alpha) + 0.006 (1 - \alpha)^2} = \frac{\alpha}{\alpha + 0.6 (1 - \alpha)} = \frac{\alpha}{0.4 \alpha + 0.6}$

-------------------

**When are the two probabilities equal?**

$2 \alpha - \alpha^2 = \frac{\alpha}{0.4 \alpha + 0.6}$

$2 - \alpha = \frac{1}{0.4 \alpha + 0.6}$

$(2 - \alpha)(0.4 \alpha + 0.6) = 1$

$0.8 \alpha + 1.2 - 0.4 \alpha^2 - 0.6 \alpha - 1 = 0$

$0.4 \alpha^2 - 0.2 \alpha - 0.2 = 0$

$2 \alpha^2 - \alpha - 1 = 0$

$\frac{1 ^+_{-} \sqrt{1 + 8}}{4} = 1$ or $-0.5 \Rightarrow$ **The probability without evidence is always greater than with evidence, as illustrated on the figure below. Therefore, the data provides evidence against the proposition that one of the animals is a kingfisher.**

```python
def kingfisher_prior(alpha):

    return 2 * alpha - alpha ** 2


def kingfisher_posterior(alpha):

    return alpha / (0.4 * alpha + 0.6)
```

```python
plt.plot(alphas, kingfisher_prior(alphas), label="without evidence")
plt.plot(alphas, kingfisher_posterior(alphas), label="with evidence")
plt.xlabel("alpha", fontsize="x-large")
plt.ylabel("probability of kingfisher", fontsize="x-large")
plt.legend(fontsize="x-large")
```
